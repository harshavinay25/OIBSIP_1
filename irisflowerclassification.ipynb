DIFFERENTIATING THE FEATURES AND TARGET

```python
data=df.values
X=data[:,0:4]
Y=data[:,4]
print(X)
print(Y)
```

TRAIN AND TEST THE MODEL

```python
# install scikit-learn package

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.svm import SVC
svn = SVC()
svn.fit(X_train, y_train)
```

```python
predict = svn.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict)
```

```python
from numpy import random
x = random.randint(10,size=(3,4)) #used random function for random generation of values to test the model
print(x)
```

```python
predict = svn.predict(x)
print(predict)
```

LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
m = LogisticRegression()
m.fit(X_train, y_train)
prediction = m.predict(X_train)
print(accuracy_score(y_train,prediction))
```
